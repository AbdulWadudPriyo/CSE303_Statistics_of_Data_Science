```python
x = df["tem"].mean()
df["tem"].fillna(x, inplace = True)
df
```

```python
y = df["rain"].median()
df["rain"].fillna(y, inplace = True)
df
```

```python
if df["Month"].isna().any():
   p = df["Month"].mode()[0]
df["Month"].fillna(p, inplace = True)

if df["Year"].isna().any():
   q = df["Year"].mode()[0]
df["Year"].fillna(q, inplace = True)
df
```
