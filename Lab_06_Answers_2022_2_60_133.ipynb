```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error,classification_report
from sklearn.linear_model import LinearRegression
```

Check for any missing (null) values in the dataset (if needed).

```python
print(df.isnull().values.any())
print(df.isnull().sum())
```

```python
df['Education'] = df['Education'].fillna('Unknown')
df['Training_Hours'] = df['Training_Hours'].fillna(0)
df['Experience'] = df['Experience'].fillna(0)
df
```

Perform appropriate mapping if needed

```python
education_map = {'Bachelor': 0, 'Master': 1, 'PhD': 2, 'Unknown': 3}
department_map = {'Sales': 0, 'Engineering': 1, 'HR': 2}
df['Education'] = df['Education'].map(education_map)
df['Department'] = df['Department'].map(department_map)
df
```

Run linear regression model on the dataset

```python
x = df.drop(columns = ['Performance_Score'])
y = df['Performance_Score']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("x_train\n",x_train)
print("y_train\n",y_train)
print("x_test\n",x_test)
print("y_test\n",y_test)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_pred
```

```python
lr_mae = mean_absolute_error(y_test, lr_pred)
lr_mse = mean_squared_error(y_test, lr_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))

print('MAE Linear Regression: ', lr_mae)
print('MSE Linear Regression: ', lr_mse)
print('RMSE Linear Regression: ', lr_rmse)
```
